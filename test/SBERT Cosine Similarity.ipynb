```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity


model = SentenceTransformer('all-MiniLM-L6-v2')


file_path = 'task 4.xlsx'  
df = pd.read_excel(file_path)


if df.shape[1] < 2:
    raise ValueError("Excel 文件至少需要两列")

# 4. 
similarities = []
for i, row in df.iterrows():
    text1 = str(row[0])  
    text2 = str(row[1])  

    emb1 = model.encode(text1)
    emb2 = model.encode(text2)

    sim = cosine_similarity([emb1], [emb2])[0][0]
    similarities.append(sim)


df['Cosine Similarity'] = similarities


df.to_excel('similarity_output.xlsx', index=False)


print("完成 ✅ 相似度已保存到 similarity_output.xlsx")
```

```python
import pandas as pd
from sentence_transformers import SentenceTransformer
import numpy as np

# 加载模型
model = SentenceTransformer('all-MiniLM-L6-v2')

# 读取 Excel 文件（假设前两列是文本）
df = pd.read_excel("task 4.xlsx")  # 替换为你的文件名

# 检查列数
if df.shape[1] < 2:
    raise ValueError("Excel 文件至少需要两列文本")

# 创建存储结果的列表
data = []

for idx, row in df.iterrows():
    text1 = str(row[0])
    text2 = str(row[1])

    vec1 = model.encode(text1)
    vec2 = model.encode(text2)

    # 你可以保存所有 384 维，也可以只保存前 5 维
    row_result = {
        'Text1': text1,
        'Text2': text2,
        'Vec1_Dim1': vec1[0], 'Vec1_Dim2': vec1[1], 'Vec1_Dim3': vec1[2], 'Vec1_Dim4': vec1[3], 'Vec1_Dim5': vec1[4],
        'Vec2_Dim1': vec2[0], 'Vec2_Dim2': vec2[1], 'Vec2_Dim3': vec2[2], 'Vec2_Dim4': vec2[3], 'Vec2_Dim5': vec2[4],
    }
    data.append(row_result)

# 转为 DataFrame
result_df = pd.DataFrame(data)

# 保存为 Excel
result_df.to_excel("output_vectors.xlsx", index=False)

print("✅ 完成：前两列文本的向量已写入 output_vectors.xlsx")
```

```python
#example: 
import numpy as np

def manual_cosine_similarity(a, b):
    a = np.array(a)
    b = np.array(b)
    dot = np.dot(a, b)
    norm_a = np.linalg.norm(a)
    norm_b = np.linalg.norm(b)
    return dot / (norm_a * norm_b)

# 示例
A = [1, 2, 3]
B = [4, 5, 6]
print(manual_cosine_similarity(A, B))  # 输出约 0.9746
```
